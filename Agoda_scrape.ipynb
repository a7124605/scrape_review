```python
import requests
import json
#點選「更多評論」後的網址
link = "https://www.agoda.com/zh-hk/caesar-park-hotel-kenting/reviews/kenting-tw.html?finalPriceView=1&isShowMobileAppPrice=false&cid=1896242&numberOfBedrooms=&familyMode=false&adults=1&children=0&rooms=1&maxRooms=0&checkIn=2023-06-7&isCalendarCallout=false&childAges=&numberOfGuest=0&missingChildAges=false&travellerType=-1&showReviewSubmissionEntry=false&currencyCode=TWD&isFreeOccSearch=false&tag=48a67f72-df63-4a3d-9418-9647b9f51f01&isCityHaveAsq=false&tspTypes=16&los=1&searchrequestid=e3a716b0-3145-4031-9e8d-f53dbac6c07e"
browser.get(link)
time.sleep(2)

```

```python
#Scroll down the page in Agoda
#Click "Readmore"

import time
from selenium.common.exceptions import NoSuchElementException


click_more_count = 0
count = 0


def click_more_review():
    global click_more_count
    button = browser.find_element(By.XPATH,'//*[@id="reviewSection"]/div[5]/button')
    try:
        browser.execute_script("arguments[0].click();", button)
        click_more_count += 1
        print(f"more review{click_more_count}times")
        time.sleep(1)
    except NoSuchElementException:
        raise NoSuchElementException 
    
    

while True:
    
    last_height = browser.execute_script("return document.body.scrollHeight")  # checking page height in pixels before scrolling
    browser.execute_script("window.scrollBy(0, document.body.scrollHeight)")                        # scroll down 10000 pixels
    #browser.execute_script("window.scrollBy(0, -200)")                         # scroll up by 200 pixels (if this is not done, new data will not be loaded)
    time.sleep(2)
    new_height = browser.execute_script("return document.body.scrollHeight")   # check page height after scrolling
    
    try:
        click_more_review()
        #count = 0        
    except NoSuchElementException:
        print("ScrollDown_Page_finish")
        break


```
